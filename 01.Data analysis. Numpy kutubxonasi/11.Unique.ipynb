```setdiff1d``` haqida

````python
# data
names1 = np.array(['Jasur', 'Abdurahmon', 'Hasan', 'Muhammad', 'Sarvar', 'Temur'])
names2 = np.array(['Sarvar', 'Abdurahmon', 'Hasan', 'Temur', 'Sarvar', 'Temur'])

# names1 dagi elementlarning qaysilari names2 da yo'qligini topish
ayirma = np.setdiff1d(names1, names2)

print(ayirma)
````

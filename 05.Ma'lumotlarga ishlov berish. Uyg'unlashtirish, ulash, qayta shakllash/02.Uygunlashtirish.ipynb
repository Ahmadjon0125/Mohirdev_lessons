### Sodda misollar

````python
A = pd.DataFrame({
    'Transport': ['poezd', 'avtomobil', 'kema'],
    'Tezligi': ['70', '180', '45']
})

A
````

````python
B = pd.DataFrame({
    'Transport': ['poezd', 'avtomobil', 'tramvay'],
    'Narxi': ['98000', '10000', '65000']
})

B
````

### merge

````python
merged = pd.merge(A,B) # inner join
merged
````

````python
merged_outer = pd.merge(A,B, how='outer') # outer join
merged_outer
````

````python
merged_left = pd.merge(A,B, how='left') # left join
merged_left # faqat left tomondan umumiy bo'lmaganlarini oladi
````

````python
merged_right = pd.merge(A,B, how='right') # right join
merged_right # faqat right tomondan umumiy bo'lmaganlarini oladi
````

````python
A = pd.DataFrame({
    'Transport_turi': ['poezd', 'avtomobil', 'kema'],
    'Tezligi': ['70', '180', '45']
})

A
````

````python
B
````

````python
merg = pd.merge(A,B, left_on='Transport_turi', right_on='Transport')
merg
````

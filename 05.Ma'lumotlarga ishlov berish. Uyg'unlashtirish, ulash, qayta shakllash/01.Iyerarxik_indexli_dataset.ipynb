### Dunyo shaharlarining havo va suv sifati ma'lumotlari ustida amallar bajarish

````python
!wget https://raw.githubusercontent.com/MansurCompAI/Practicum_Datasets/refs/heads/master/world_cities_air_water_quality.csv
````

````python
df = pd.read_csv('world_cities_air_water_quality.csv')
df
````

````python
df.info()
````

````python
columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']
df.columns = columns
df
````

````python
df['Region'] = df['Region'].map(lambda x: x.replace('"', ''))
````

````python
df['Country'] = df['Country'].map(lambda x: x.replace('"', ''))
````

````python
df
````

````python
df.to_csv('updated_dataset.csv', index=False)
````

````python
df[df['Country'] == 'Germany']
````

````python
df['Country'] = df['Country'].map(lambda x: x.strip())
df
````

````python
df[df['Country'] == 'Germany']
````

````python
df[df['Country'] == 'Uzbekistan']
````

### ```pivot table``` orqali iyerarxik indexlash

````python
df
````

````python
df1 = pd.pivot_table(df, index=['Country', 'City'])
df1
````

````python
df1 = pd.pivot_table(df, 
                     index=['Country', 'City'], 
                     values=['AirQuality', 'WaterPollution'])
df1
````

````python
df1.loc[['Uzbekistan', 'Germany']]
````

### Ustunlarni iyerarxik ko'rinishga o'tkazish

````python
df2 = df1.copy()
df2
````

````python
df2['AirPollution'] = 100 - df2['AirQuality']
df2['WaterQuality'] = 100 - df2['WaterPollution']
df2
````

````python
df3 = pd.DataFrame(df2.values, 
                  index = df2.index, 
                  columns= [['AIR', 'AIR', 'WATER', 'WATER'], 
                           ['AirQuality', 'AirPollution', 'WaterQuality', 'WaterPollution']])
df3
````

````python
df3['AIR']
````

````python
df3.columns.names = ["Object", "Info"]
df3
````

### ```swaplevel```

````python
df3.swaplevel('Country', 'City') # Iyerarxik indexlarni o'rnini almashtirish
````

```python
import pandas as pd
import sqlite3 as sql
from pprint import pprint as print
```

```python
import urllib.request
import shutil

url = "https://www.sqlitetutorial.net/wp-content/uploads/2018/03/chinook.zip"

# Brauzer ekanligimizni bildirish uchun 'User-Agent' qo'shamiz
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
urllib.request.install_opener(opener)

try:
    print("Yuklanmoqda...")
    urllib.request.urlretrieve(url, "chinook.zip")
    
    print("Arxivdan chiqarilmoqda...")
    shutil.unpack_archive("chinook.zip")
    print("Muvaffaqiyatli yakunlandi!")
    
except Exception as e:
    print(f"Xatolik yuz berdi: {e}")
```

```python
db = sql.connect('chinook.db')
cursor = db.cursor()
```

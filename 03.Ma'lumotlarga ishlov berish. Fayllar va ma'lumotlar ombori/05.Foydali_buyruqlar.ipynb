```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query,db)
countries
```

```python
query = "SELECT * FROM elect WHERE country ='Uzbekistan' AND year > 2010"
dfuz = pd.read_sql(query, db)
dfuz
```

```python
query = "SELECT country,year, coal,gas, hydro, oil FROM elect WHERE country ='Uzbekistan' AND year BETWEEN 2010 AND 2015"
pd.read_sql(query,db)
```

```python
query = "SELECT country,year, coal,gas, hydro, oil FROM elect WHERE country ='Uzbekistan' AND year IN (2005,2008, 2009,2015)"
pd.read_sql(query,db)
```

```python
# LIKE operatori uchun namunalar:
# %var   -> Anvar, Sarvar (oxiri 'var' bilan tugaydigan har qanday so'z)
# _ar    -> car, bar, tar (bitta ixtiyoriy belgi + 'ar')
# __ar   -> star (ikkita ixtiyoriy belgi + 'ar')
# %ol%   -> bola, xola, bolalar ('ol' birikmasi so'zning istalgan joyida bor bo'lsa)

query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN (2019, 2020, 2021)"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year=2020"
pd.read_sql(query, db)
```

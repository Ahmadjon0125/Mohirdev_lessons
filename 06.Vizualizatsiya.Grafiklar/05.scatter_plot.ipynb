```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="year", y="price")
plt.show()
```

```python
df = df[df.year > 2000]
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="year", y="price")
plt.show()
```

```python
df = df[df.year > 2000]
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x="year", y="price")
plt.show()
```

```python
df = df[df.year > 2000]
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="mileage", y="price", hue="transmission", s=100)
plt.show()
```

```python
df = df[df.year > 2000]
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="mileage", y="price", hue="transmission", style="fuelType", s=100)
plt.show()
```

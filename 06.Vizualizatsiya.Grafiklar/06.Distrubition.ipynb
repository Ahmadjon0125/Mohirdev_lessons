```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Ma'lumotlarni o'qish
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")

# 1. Faqat 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df = df[df.year >= 2010] 

# 2. Transmission ustunidan 'Other' (boshqa) qiymatlarini olib tashlaymiz
# Faqat avtomat, mexanik va yarim-avtomatlar qoladi
df = df[df.transmission != 'Other'] 

# 3. fuelType ustunidan ham 'Other' qiymatlarini filtrlaymiz
# Faqat benzin, dizel va gibridlar qoladi
df = df[df.fuelType != 'Other'] 

# Natijani tekshirish
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.displot(data=df, x="price", hue="transmission", height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x="price", hue="transmission", multiple="stack", height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x="price", col="transmission",  height=5, aspect=2) ## hue o'rniga col ishlatdik multiple="stack" ni olib tashladik
plt.show()
# 3 ta grafikni chizib beradi
```

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.show()  
```

```python
sns.displot(data=df, x='price', kde=True , height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='price', shade=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='price', hue='transmission', shade=True)
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', kind='kde', height=5, aspect=2)
plt.xticks(rotation=45)
plt.show()
```

CDF

```python
sns.displot(data=df, x='price', kind='ecdf', height=5, aspect=2)
plt.xticks(rotation=45)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```

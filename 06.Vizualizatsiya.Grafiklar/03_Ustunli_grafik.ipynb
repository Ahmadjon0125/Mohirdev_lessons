```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv", index_col=0)
df.columns = ['maydon', 'aholi']
df
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x=df.index, y=df['aholi'])
plt.title("O'zbekiston viloyatlarining aholi soni")
plt.xlabel("Viloyatlar")
plt.ylabel("Aholi soni")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x=df.index, y=df['maydon'])
plt.title("O'zbekiston viloyatlarining maydoni")
plt.xlabel("Viloyatlar")
plt.ylabel("Maydon (km²)")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

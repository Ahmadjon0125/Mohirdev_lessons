```python
plt.figure(figsize=(16, 6))

x = np.arange(0, 4* np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y, linewidth=3, color='blue', linestyle=':' , marker='o', markersize=10, markerfacecolor='red', markeredgecolor='black')
plt.title('sin(x) funksiya')
plt.xlabel('pi')
plt.ylabel('y=sin(x)')
plt.grid()
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x) qiymatlari
cos = np.cos(x) # cos(x) qiymatlari

# Grafik maydonini yaratish (hajmi 12x6)
plt.figure(figsize=(12, 6))

# sin(x) grafigini chizish
plt.plot(x, sin, 
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# cos(x) grafigini chizish
plt.plot(x, cos, 
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# Grafik nomi
plt.title('sin(x) va cos(x) funksiya grafigi')

# x-o'q nomi
plt.xlabel("pi")

# Har bir chiziq haqida ma'lumot (Legend)
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')

# Setkani (grid) sozlash
plt.grid(color='tomato', alpha=0.5, linestyle='--')
#saqlash
plt.savefig('sin_cos_grafik.png')

# Grafikni ko'rsatish
plt.show()
```

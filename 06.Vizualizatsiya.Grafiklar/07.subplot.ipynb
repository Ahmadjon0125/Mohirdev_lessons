```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# 1. Ma'lumotlarni GitHub-dan o'qib olamiz
url = "https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv"
df = pd.read_csv(url)

# 2. Ma'lumotlarni filtrlash (tozalash)
# 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df = df[df.year >= 2010] 

# 'transmission' ustunidan 'Other' (noma'lum) qiymatlarni olib tashlaymiz
df = df[df.transmission != 'Other'] 

# 'fuelType' ustunidan 'Other' (noma'lum) qiymatlarni olib tashlaymiz
df = df[df.fuelType != 'Other'] 

# 3. Natijani tekshirish
df.head()
```

```python

x = np.arange(0, 4 * np.pi, 0.1) # 0 dan 4*pi gacha 0.1 qadam bilan
sin = np.sin(x) # sin(x) qiymatlari
cos = np.cos(x) # cos(x) qiymatlari

plt.figure(figsize=(12, 6))

# 2. sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD', # Siyohrang (Hex kodda)
         alpha=0.7,
         marker='h') # h - hexagon (olti burchak) markeri
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # Setkani sozlash

# 3. cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c', # Cyan (och ko'k)
         alpha=0.7,
         marker='v') # v - pastga qaragan uchburchak markeri
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # Setkani sozlash

plt.show()
```

```python

x = np.arange(0, 4 * np.pi, 0.1) # 0 dan 4*pi gacha 0.1 qadam bilan
sin = np.sin(x) # sin(x) qiymatlari
cos = np.cos(x) # cos(x) qiymatlari

plt.figure(figsize=(12, 6))

# 2. sin(x) grafigi
plt.subplot(2, 1, 1) # 2 qator, 1 ustun, 1-chi grafik
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD', # Siyohrang (Hex kodda)
         alpha=0.7,
         marker='h') # h - hexagon (olti burchak) markeri
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # Setkani sozlash

# 3. cos(x) grafigi
plt.subplot(2, 1, 2) # 2 qator, 1 ustun, 2-chi grafik
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c', # Cyan (och ko'k)
         alpha=0.7,
         marker='v') # v - pastga qaragan uchburchak markeri
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # Setkani sozlash

plt.show()
```

Hisobotlar bugungi kunda dunyodagi baxtiyorlik darajasining holatini koʻrsatadi.

### **Ustunlar tavsifi**  

1. **Country** – mamlakat nomi  
   - Bu satrda mamlakatlarning nomlari joylashgan.  

2. **Happiness.Rank** – baxtlilik darajasi bo‘yicha reyting  
   - Har bir mamlakatning baxt darajasi bo‘yicha egallagan o‘rni.  
   - **1** – eng baxtli mamlakat, **so‘nggi o‘rin** – eng baxtsiz mamlakat.  

3. **Happiness.Score** – baxtlilik darajasi  
   - Mamlakatning umumiy baxtlilik darajasi **(0–10 oralig‘ida)**.  
   - Bu baho turli omillarga asoslangan holda berilgan.  

4. **Whisker.high** – yuqori ishonch oralig‘i (Upper Confidence Interval)  
   - **Happiness Score** qiymatining yuqori chegarasi (ishonchlilik darajasi bilan).  

5. **Whisker.low** – pastki ishonch oralig‘i (Lower Confidence Interval)  
   - **Happiness Score** qiymatining pastki chegarasi.  

6. **Economy..GDP.per.Capita.** – iqtisodiy farovonlik (YAIM)  
   - Aholi jon boshiga to‘g‘ri keladigan yalpi ichki mahsulot (YAIM).  
   - Mamlakat iqtisodining baxtlilik darajasiga ta’sirini ko‘rsatadi.  

7. **Family** – jamiyatning qo‘llab-quvvatlashi  
   - Ijtimoiy aloqalar va oilaviy qo‘llab-quvvatlash darajasi.  
   - Ushbu qiymat qanchalik yuqori bo‘lsa, odamlar o‘zini shunchalik baxtli his qiladi.  

8. **Health..Life.Expectancy.** – sog‘lom umr davomiyligi  
   - Odamlarning o‘rtacha qancha umr ko‘rishi va sog‘lom hayot kechirishi darajasi.  
   - Tibbiy xizmat sifati va hayot sharoitlari bilan bog‘liq.  

9. **Freedom** – hayot tanlovlarida erkinlik  
   - Odamlarning o‘z hayotini erkin boshqarish darajasi.  
   - Mamlakatdagi demokratik va fuqarolik erkinliklarini aks ettiradi.  

10. **Generosity** – saxiylik darajasi  
   - Jamiyatdagi saxiylik va xayriya qilishga moyillik darajasi.  
   - Ushbu qiymat qanchalik yuqori bo‘lsa, odamlar ko‘proq yordam berishga tayyor bo‘ladi.  

11. **Trust..Government.Corruption.** – korrupsiya haqidagi tasavvurlar  
   - Mamlakatdagi korrupsiya darajasi bo‘yicha odamlarning fikri.  
   - **0 ga yaqin bo‘lsa**, odamlar davlatga ishonmaydi.  
   - **Yuqori qiymat bo‘lsa**, davlatga ishonch baland bo‘ladi.  

12. **Dystopia.Residual** – baxtiyorlik darajasiga ta’sir etmaydigan omillar  
   - Ideal bo‘lmagan (Dystopia) sharoitlarda baxtlilik darajasini oshirishga yordam beruvchi faktorlar.  
   - Bu ko‘rsatkich mamlakat baxtlilik darajasi **modelining qoldiq qiymati** sifatida tushuniladi.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/happines_rank_2017.csv"
df = pd.read_csv(url)
df.head()
```

#1) Baxtlilik darajasi va aholi jon boshiga to‘g‘ri keladigan YAIM o‘rtasidagi bog‘liqlikni scatterplot bilan vizualizatsiya qiling.

```python
#javobni shu yerda yozing.
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x="Economy..GDP.per.Capita.",
    y="Happiness.Score",
    s=100,
    alpha=0.7,
    color='teal'
)

# 3. Grafik bezaklari
plt.title("YAIM va Baxtlilik darajasi o'rtasidagi bog'liqlik", fontsize=14)
plt.xlabel("YAIM (Aholi jon boshiga)", fontsize=12)
plt.ylabel("Baxtlilik darajasi (Score)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

#2) Baxtlilik darajasi (`Happiness.Score`) taqsimotini histogram orqali chizing.

```python
#javobni shu yerda yozing.
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Happiness.Score', kde=True, color='skyblue')

# Bezaklar
plt.title('Dunyo davlatlari bo\'yicha baxtlilik darajasi taqsimoti', fontsize=14)
plt.xlabel('Baxtlilik darajasi (Score)', fontsize=12)
plt.ylabel('Davlatlar soni', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

#3) Eng baxtli va eng baxtsiz 10 ta mamlakatni barplot orqali ko‘rsating.

```python
#javobni shu yerda yozing.
top_10 = df.nlargest(10, 'Happiness.Score')
bottom_10 = df.nsmallest(10, 'Happiness.Score')

# 3. Ikkita grafik uchun maydon tayyorlaymiz (1 qator, 2 ustun)
fig, ax = plt.subplots(1, 2, figsize=(18, 8))

# 4. Chap tomonda: Eng baxtli 10 ta davlat (Yashil ranglar palitrasi)
sns.barplot(data=top_10, x='Happiness.Score', y='Country', ax=ax[0], palette='Greens_r')
ax[0].set_title('Eng baxtli 10 mamlakat', fontsize=14)
ax[0].set_xlabel('Baxtlilik darajasi')
ax[0].set_ylabel('Mamlakat')

# 5. O'ng tomonda: Eng baxtsiz 10 ta davlat (Qizil ranglar palitrasi)
sns.barplot(data=bottom_10, x='Happiness.Score', y='Country', ax=ax[1], palette='Reds')
ax[1].set_title('Eng baxtsiz 10 mamlakat', fontsize=14)
ax[1].set_xlabel('Baxtlilik darajasi')
ax[1].set_ylabel('') # Ikkinchi grafikda y-o'qi nomini olib tashlaymiz

# 6. Grafiklarni tekshirib, chiroyli holatga keltiramiz
plt.tight_layout()
plt.show()
```

#4)  Baxtlilik darajasi (`Happiness.Score`) va ijtimoiy omillar (Freedom, Family Support, Generosity) bilan scatterplot (`subplot` orqali ketma ket) chizing!

```python
#javobni shu yerda yozing.
# 3. Grafik maydonini yaratamiz (1 qator, 3 ustun)
fig, ax = plt.subplots(1, 3, figsize=(16, 6))

# Birinchi grafik: Family va Happiness
sns.scatterplot(data=df, x='Family', y='Happiness.Score', ax=ax[0], color='blue', alpha=0.7)
ax[0].set_title('Family vs Happiness')
ax[0].grid(True, linestyle='--', alpha=0.5)

# Ikkinchi grafik: Freedom va Happiness
sns.scatterplot(data=df, x='Freedom', y='Happiness.Score', ax=ax[1], color='green', alpha=0.7)
ax[1].set_title('Freedom vs Happiness')
ax[1].grid(True, linestyle='--', alpha=0.5)

# Uchinchi grafik: Generosity va Happiness
sns.scatterplot(data=df, x='Generosity', y='Happiness.Score', ax=ax[2], color='orange', alpha=0.7)
ax[2].set_title('Generosity vs Happiness')
ax[2].grid(True, linestyle='--', alpha=0.5)

# Grafiklarning bir-biriga yopishib qolmasligi uchun
plt.tight_layout()
plt.show()
```

#6) Baxtlilik darajasi va YAIM o‘rtasidagi bog‘liqlikni chiziqli regressiya orqali tekshiring!

```python
#javobni shu yerda yozing.
plt.figure(figsize=(10, 6))

# 3. Chiziqli regressiya grafigini chizamiz
sns.regplot(
    data=df,
    x="Economy..GDP.per.Capita.",
    y="Happiness.Score",
    line_kws={"color": "red"},   # Regressiya chizig'ini qizil qilamiz
    scatter_kws={"alpha": 1}   # Nuqtalarni biroz shaffof qilamiz
)

# 4. Bezaklar
plt.title("YAIM va Baxtlilik: Chiziqli Regressiya tahlili", fontsize=14)
plt.xlabel("YAIM (Iqtisodiy ko'rsatkich)", fontsize=12)
plt.ylabel("Baxtlilik darajasi (Score)", fontsize=12)
plt.grid(True, alpha=1)

plt.show()
```

#7) Issiqlik xaritasi (heatmap) orqali korrelyatsiyalarni ko‘rsating.

```python
#javobni shu yerda yozing.
# 2. Faqat sonli ustunlar orasidagi korrelyatsiyani hisoblaymiz
# (Country kabi matnli ustunlarni hisobga olmaymiz)
corr_matrix = df.corr(numeric_only=True)

# 3. Grafik o'lchami
plt.figure(figsize=(12, 10))

# 4. Heatmap chizamiz
sns.heatmap(data=corr_matrix,
            annot=True,      # Katakchalar ichiga raqamlarni yozish
            cmap='coolwarm', # Ranglar palitrasi (sovuqdan issiqqa)
            fmt=".2f",       # Verguldan keyin 2 ta raqam ko'rsatish
            linewidths=0.5)  # Kataklar orasidagi chiziq qalinligi

plt.title('Baxtlilik darajasi va omillar o\'rtasidagi korrelyatsiya xaritasi', fontsize=16)
plt.show()
```

#11) Baxtlilik darajasi bo‘yicha mamlakatlarni “Juda baxtli”, “Baxtli”, “O‘rtacha”, “Baxtsiz” toifalariga ajrating va har bir guruh uchun o‘rtacha iqtisodiy ko‘rsatkichlarni (GDP per Capita) ko‘rsating.

```python
#javobni shu yerda yozing.

# 2. Toifalash uchun chegaralar (bins) va nomlar (labels) yaratamiz
# Baxtlilik darajasi odatda 2.5 dan 7.5 gacha. Shunga qarab chegaralaymiz:
bins = [0, 4, 5.5, 6.5, 10]
labels = ["Baxtsiz", "O‘rtacha", "Baxtli", "Juda baxtli"]

# 3. Yangi 'Status' ustunini yaratamiz
df['Status'] = pd.cut(df['Happiness.Score'], bins=bins, labels=labels)

# 4. Har bir guruh uchun YAIM (GDP) o'rtacha qiymatini hisoblaymiz
status_gdp = df.groupby('Status')['Economy..GDP.per.Capita.'].mean().reset_index()

# 5. Vizualizatsiya (Barplot)
plt.figure(figsize=(8, 4))
sns.barplot(data=status_gdp, x='Status', y='Economy..GDP.per.Capita.', palette='magma')

# 6. Bezaklar
plt.title('Baxtlilik toifalari bo\'yicha o\'rtacha YAIM ko\'rsatkichi', fontsize=14)
plt.xlabel('Baxtlilik holati', fontsize=12)
plt.ylabel('O\'rtacha YAIM (GDP per Capita)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

#12) Muhim o‘zgaruvchilar (GDP, Life Expectancy, Freedom, Generosity, Happiness Score) bo‘yicha pairplot orqali umumiy taqsimotni vizualizatsiya qiling.

```python
#javobni shu yerda yozing.
# 2. Muhim o'zgaruvchilarni tanlab olamiz (nomlarni to'g'rilab yozish shart)
selected_columns = [
    'Happiness.Score',
    'Economy..GDP.per.Capita.',
    'Health..Life.Expectancy.',
    'Freedom',
    'Generosity'
]

# 3. Pairplot chizamiz
# kind='reg' - har bir scatterplotga chiziqli regressiya chizig'ini ham qo'shadi
sns.pairplot(df[selected_columns], kind='reg', diag_kind='kde', plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.4}})

plt.suptitle('Muhim o‘zgaruvchilar bo‘yicha umumiy taqsimot (Pairplot)', y=1.02, fontsize=16)
plt.show()
```

#1) Quyida berilgan datasetni yuklab olib, ustun nomlarini koʻring va ularni maʼno jihatdan tushunarli shaklga keltiring. `Unnamed: 0` ustunini tashlab yuboring!

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/restaurant_sales.csv"
df = pd.read_csv(url)
df.head()
```

```python
#javobni shu yerda yozing
df.drop('Unnamed: 0', axis=1, inplace=True)
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
df
```

#2) Maʼlumotlarning umumiy holatini (info(), describe()) tekshiring.

```python
#javobni shu yerda yozing
df.info()
```

```python
df.describe()
```

#3) Har bir ustundagi boʻsh qiymatlarni aniqlang va ularni toʻldirish strategiyasini ishlab chiqing.

```python
#javobni shu yerda yozing
df.isnull().sum()
```

```python
df['category']= df['category'].fillna('Unknown')
df.isnull().sum()
```

```python
df.item = df.item.fillna("Unknown")
df.isnull().sum()
```

```python
df.quantity = df.quantity.fillna(df.quantity.median())
df.isnull().sum()
```

```python
df.order_total = df.order_total.fillna(df.quantity * df.price)
df.isnull().sum()
```

```python
mode_payment = df.payment_method.mode()[0]
df.payment_method = df.payment_method.fillna(mode_payment)
df.isnull().sum()
```

#4) Har bir ustundagi takroriy qiymatlarni toping va ularga qanday oʻzgartirish kiritish kerakligini hal qiling.

```python
#javobni shu yerda yozing
df.duplicated().sum()
```

```python
df[df.duplicated(keep=False)].sort_values(by='order_id')
```

```python
id_duplicates = df['order_id'].duplicated().sum()
id_duplicates
```

```python

text_columns = ['category', 'item', 'payment_method']

for col in text_columns:
    df[col] = df[col].astype(str).str.lower().str.strip()

print(f"Tozalashdan keyingi to'liq dublikatlar: {df.duplicated().sum()}")
```

#5) Maʼlumotlar toʻplamidagi noto‘g‘ri maʼlumotlar (masalan, son turishi kerak bo‘lgan ustunda matn bo‘lsa) mavjudligini tekshiring va to‘g‘rilang.

```python
df.head()
```

```python
#javobni shu yerda yozing
df.info()
```

```python
df.describe()
```

```python
df.price = df.price.mask(df.price <= 0, df.price.median())
df.order_total = df.order_total.mask(df.order_total <= 0, df.order_total.median())
df.describe()
```

#6) Maʼlumotlar toʻplamidagi **katta/kichik harf** muammolari boʻlgan joylarni toping va bir xil formatga keltiring.

```python
#javobni shu yerda yozing
for col in ['category', 'item', 'payment_method']:
    df[col] = df[col].astype(str).str.lower().str.strip()
df.head()
```

#7) Maʼlumotlar turini oʻzgartirish.
 * Ustunlarning maʼlumot turlarini tekshiring. Notoʻgʻri turlarga ega boʻlgan ustunlarni toʻgʻri formatga oʻtkazing.
 * `order_date` yoki unga oʻxshash ustunni datetime formatiga oʻtkazing.

```python
#javobni shu yerda yozing
df.info()
```

```python
df['order_id'] = df['order_id'].str.replace('ORD_', '', regex=False).astype('Int64')
df['customer_id'] = df['customer_id'].str.replace('CUST_', '', regex=False).astype('Int64')
df['order_date'] = pd.to_datetime(df['order_date'])
df['quantity'] = df['quantity'].astype('int64')
df['category'] = df['category'].astype('category')
df['payment_method'] = df['payment_method'].astype('category')
df.head()
```

#8) Mahsulot nomlarida noto‘g‘ri yoki nomaʼlum belgilar (?, NaN, #) mavjudligini tekshiring va ularni tuzating.

```python
#javobni shu yerda yozing
# Faqat harf va raqam bo'lmagan belgilarni o'z ichiga olgan nomlarni ko'rish
import re
pattern = r'[^a-zA-Z0-9\s]' # Harf, son va bo'shliqdan boshqa hamma narsa
strange_items = df[df['item'].str.contains(pattern, na=False)]['item'].unique()
print(strange_items)
```

```python
df['item'] = df['item'].str.replace(r'[#?]', '', regex=True).str.strip()
```

```python
df['item'].str.lower().unique()
```

#**Maʼlumotlarni filtrlash va qayta ishlash**

#9) Manfiy narx yoki 0 ga teng narx qiymatlarini toping va ularni qanday oʻzgartirish kerakligini hal qiling.

```python
#javobni shu yerda yozing
df.price = df.price.mask(df.price <= 0, df.price.median())
df.price.value_counts()
```

#10) Boʻsh toifalar (kategoriya) mavjudligini tekshiring va toʻgʻri kategoriya berish strategiyasini ishlab chiqing.

```python
#javobni shu yerda yozing
df.head()
```

```python
# Avval 'unknown' qiymatlarni NaN ga qaytaramiz (fillna ishlashi uchun)
import numpy as np
df['category'] = df['category'].replace('unknown', np.nan)

# Mahsulot nomi (item) bo'yicha kategoriyalarni guruhlab to'ldiramiz
df['category'] = df['category'].fillna(df.groupby('item')['category'].transform('first'))
df.category.value_counts()
```

#11) Bir xil, lekin turlicha yozilgan qiymatlar (Coffee, cofee, COFEE) mavjudligini tekshiring va tuzating.

```python
#javobni shu yerda yozing
df['item'] = df['item'].str.replace(r'\s+', ' ', regex=True)
df['item'] = df['item'].str.lower().str.strip()
sorted(df['item'].unique())
```

```python
df.item.value_counts().sort_index()
```

```python
df = df[df['item'] != 'unknown']
df.item.value_counts().sort_index()
```

#12) `Customer ID` yoki boshqa identifikatorlarning takrorlanishini tekshiring.

```python
#javobni shu yerda yozing
duplicate_rows = df.duplicated().sum()
if duplicate_rows > 0:
    df = df.drop_duplicates()

```

```python
unique_customers = df['customer_id'].nunique()
unique_customers
```

#13) Mahsulotning ortiqcha taʼriflarini olib tashlang (Water (0.5), Water (1.0), Water → faqat Water qoldirish).

```python
#javobni shu yerda yozing
df['item'] = df['item'].str.replace(r'\(.*\)', '', regex=True).str.strip()
df['item'].str.lower().unique()
```

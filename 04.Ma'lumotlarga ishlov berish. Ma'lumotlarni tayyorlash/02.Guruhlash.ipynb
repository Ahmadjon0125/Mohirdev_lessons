```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.price.describe()
```

```python
bins = [0, 10000, 20000, 50000]
```

```python
range = pd.cut(df.price, bins)
```

```python
range
```

```python
range.value_counts()
```

```python
names = ['arzon', 'orta', 'qimmat']
```

```python
range = pd.cut(df.price, bins, labels=names)
range
```

```python
range.value_counts()
```

```python
range = pd.cut(df.price, 3)

range.value_counts()
```

```python
range = pd.qcut(df.price, 3) #qcut chiqadigan countlarini nisbatan teng qilib ajratarkan
range.value_counts()
```

### groupby() yordamida guruhlash

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.company.unique()
```

```python
df.company.value_counts()
```

```python
df.groupby('company').mean(numeric_only=True)
```

```python
df.groupby('company').price.max()
```

```python
df.groupby(['company','body-style']).price.mean()
```

```python
df.groupby('company').agg([min, max, np.mean]) # bu eskirgan usul ekan ishlamas ekan
```

```python
df.select_dtypes(include='number').groupby(df['company']).agg([min, max, np.mean])
```

```python
df.select_dtypes(include='number').groupby(df['company']).agg([min, max, lambda x: np.mean(x)])
```
